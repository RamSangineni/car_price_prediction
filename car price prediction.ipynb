### There are outliers in some of the features, removing outliers

```python
# List of categorical variables
categorical_columns = ['fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem']

# Function to cap outliers within each category
def cap_outliers_within_category(data, cat_col, num_col):
    categories = data[cat_col].unique()

    for category in categories:
        subset = data[data[cat_col] == category]
        Q1 = subset[num_col].quantile(0.25)
        Q3 = subset[num_col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Cap the outliers at the upper bound within each category
        data.loc[(data[cat_col] == category) & (data[num_col] > upper_bound), num_col] = upper_bound

# Cap outliers for each categorical vs. numeric variable
plt.figure(figsize=(16, 12))
for i, cat_col in enumerate(categorical_columns, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(data=df, x=cat_col, y='price')
    plt.title(f'{cat_col} vs. Price')
    plt.xticks(rotation=45)
    
    # Cap outliers for the current categorical variable
    cap_outliers_within_category(df, cat_col, 'price')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

## selecting features which are neccessary for car price prediction, discarding rest of the features 

```python
# Split the data into training and testing sets
X = df1.drop(columns=['price'])  # Features
Y = df1['price']  # Target variable
```

```python
label_encoder = LabelEncoder()
```

```python
categorical_columns = X[['fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginetype', 'cylindernumber', 'fuelsystem']]
```

```python
# Applying label encoding to each categorical column
for column in categorical_columns:
    X[column] = label_encoder.fit_transform(X[column])
```

```python
numeric_columns = X.select_dtypes(include=['int64', 'float64'])
```

```python
numeric_columns=numeric_columns[['symboling','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg']]
```

```python
# Create a StandardScaler object
scaler = StandardScaler()
```

```python
# Applying STANDARD SCALING to each numeric column
for column in numeric_columns:
    # Reshape the column to a 2D array before scaling
    X[column] = scaler.fit_transform(X[column].values.reshape(-1, 1))
```

```python
## After Encoding categorical features & scaling Numerical features
X.head(5)
```

```python
###Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
##Build and Train the Regression Model
model = Pipeline([
    ('regressor', RandomForestRegressor(random_state=42))
])
```

```python
model.fit(X_train, Y_train)
```

```python
### Evaluate the Model
Y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
```
